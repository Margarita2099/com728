```python
import pandas as pd
import csv
import list
import tables
import tui
import plot


def read_data(file_path): # function for reading data
    tui.started(f"Reading data from {file_path}")
    records = [] # made a list for values
    with open(file_path) as csv_file:
        csv_reader = csv.reader(csv_file)
        # read in and store the headings separately
        headings = next(csv_reader)
        # Display a list of values for each line
        for line in csv_reader:
            records.append(line)
    tui.completed()
    return records, headings


def run():
    records, headings = read_data("CarPrice.csv") #read data
    cars = pd.read_csv('CarPrice.csv') #read data
    tui.welcome()  # welcoming message
    #Loop executes only if the test condition is True
    while True:
        choice = tui.menu(0)
        # use multiple conditions
        if choice == 1:
            var_sub_menu = tui.menu(1)
            while var_sub_menu not in range(1,5):
                print("Please try again!")
                var_sub_menu = tui.menu(1)
            if var_sub_menu == 1:
                list.ret_by_id(records, headings)
            if var_sub_menu == 2:
                list.cylinder_number(records)
            if var_sub_menu == 3:
                list.car_body(records)
            if var_sub_menu == 4:
                list.info_by_id(records)
        elif choice == 2:
            var_sub_menu = tui.menu(2)
            while var_sub_menu not in range(1,5):
                print("Please try again!")
                var_sub_menu = tui.menu(2)
            if  var_sub_menu == 1:
                tables.alphabet(cars)
            if  var_sub_menu == 2:
                tables.price_for_car_body(cars)
            if  var_sub_menu == 3:
                tables.most_expensive_by_carbody(cars)
            if var_sub_menu == 4:
                tables.brands(cars)
        elif choice == 3:
            var_sub_menu = tui.menu(3)
            while var_sub_menu not in range(1,4):
                print("Please try again!")
                var_sub_menu = tui.menu(3)
            if  var_sub_menu == 1:
                plot.bar_chart(cars)
            if  var_sub_menu == 2:
                plot.subplot(cars)
            if  var_sub_menu == 3:
                plot.pie_subplot(cars)
        elif choice == 4:
            print("Good bye!")
            #the loop encounters a break statement and finishes
            break
        else:
            print("Please try again!")
#The function purpose is to separate the code that will be executed when the code is called as a module and when the module itself is run as a separate file.
if __name__ == "__main__":
    run()
```
